```python
import pandas as pd
import ast
import json
```

```python
dfs=[]
for i in range(2,170):
    df=pd.read_json(f'nurses_{i}.json')
    df['dates_off']=df['dates_off'].apply(lambda x: [x[i] for i in range(len(x))])
    df['len_dates_off']=df['dates_off'].apply(lambda x: len(x))
    df['len_vacations']=df['vacations'].apply(lambda x: len(x))
    df['instance']=i
    dfs.append(df)
df=pd.concat(dfs)
df.sort_values(by=['len_vacations'],ascending=False,inplace=True)
```

```python
pd.set_option('display.max_rows', None)
df.groupby(['instance','shift_preference']).agg({'len_dates_off':['mean'],'len_vacations':['mean'],'nurse_name':['count']})
```
